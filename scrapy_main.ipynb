```python
import requests
from bs4 import BeautifulSoup

amazon_url = "https://www.amazon.in/s?k=iphones"

# Send an HTTP GET request to the Amazon page
response = requests.get(amazon_url)

if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Use BeautifulSoup to locate the data you want to scrape
    prices = [price.text for price in soup.find_all('span', class_='a-price-whole')]
    ratings = [rating.text for rating in soup.find_all('span', class_='a-size-base')]

    # Zip the data into pairs
    results = list(zip(prices, ratings))

    for result in results:
        print(result)
else:
    print("Failed to retrieve the Amazon page.")

```

```python
import scrapy

class AmazonSpider(scrapy.Spider):
    name = 'amazon'
    start_urls = ['https://www.amazon.in/s?k=iphones']

    def parse(self, response):
        for product in response.css('.s-result-item'):
            yield {
                'price': product.css('.a-price-whole::text').get(),
                'ratings': product.css('.a-size-base::text').get(),
            }

        next_page = response.css('.s-pagination-next a::attr(href)').get()
        if next_page:
            yield response.follow(next_page, self.parse)
```

```python
import requests
from bs4 import BeautifulSoup

amazon_url = "https://www.amazon.in/s?k=iphones"

# Send an HTTP GET request to the Amazon page
response = requests.get(amazon_url)

if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find product containers on the page
    product_containers = soup.find_all("div", class_="s-result-item")

    for product in product_containers:
        product_name = product.find("span", class_="a-text-normal").text
        product_price = product.find("span", class_="a-price-whole").text
        customer_ratings = product.find("span", class_="a-size-base").text

        print(f"Product: {product_name}")
        print(f"Price: {product_price}")
        print(f"Customer Ratings: {customer_ratings}")
        print("-" * 50)
else:
    print("Failed to retrieve the Amazon page.")
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL you want to scrape (e.g., a YouTube search page)
youtube_url = 'https://www.youtube.com/results?search_query=python+tutorial'

# Send an HTTP GET request to the YouTube search page
response = requests.get(youtube_url)

if response.status_code == 200:
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find video titles and links on the page
    video_titles = soup.find_all('a', class_='yt-simple-endpoint style-scope ytd-video-renderer')
    
    for title in video_titles:
        video_title = title.text.strip()
        video_link = 'https://www.youtube.com' + title['href']
        print(f"Title: {video_title}")
        print(f"Link: {video_link}")
        print("-" * 50)
else:
    print("Failed to retrieve the YouTube search page.")
```
